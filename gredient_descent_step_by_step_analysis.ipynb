## Gredient Descent session 1

```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y = make_regression(n_samples = 4, n_features = 1,n_targets = 1,n_informative = 1, noise = 80, random_state = 13)
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
```

```python
# lets apply OLS
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
y_pred = reg.predict(x)
```

```python
plt.plot(x,y_pred,color="red")
plt.scatter(x,y)
plt.scatter(x,y_pred)
```

```python
# lets apply gredient descent assuming slope m is constant m = 78.35
# and lets assume starting value of intercept b is 0
y_pred = (78.35 * x).reshape(4)
y_pred
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color="red")
plt.plot(x,y_pred,color="green",label="b=0")
plt.legend()
plt.show()
```

```python
m=78.35
b=100

loss_slop = -2 * np.sum(y-m*x.ravel()-b)
loss_slop
```

```python
lr=0.1     # initilize learning rate

# now initilize step size
step_size = loss_slop*lr
step_size
```

```python
# calculating new intercept
b=b-step_size
b
```

```python
# plot the graph
y_pred2 = ((m*x)+b).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),label="OLS")
plt.plot(x,y_pred2,label="b={}".format(b))
plt.plot(x,y_pred,label="b=0",color="green")
plt.legend()
```

```python
# iteration 2
loss_slop = -2 * np.sum(y-m*x.ravel() - b)
loss_slop
```

```python
step_size = loss_slop*lr
step_size
```

```python
b2=b-step_size
b2
```

```python
# now plot graph
y_pred3 = (m*x+b).reshape(4)
y_pred3
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),label="OLs",color="red")
plt.plot(x,y_pred2,label="b = {}".format(b),color="skyblue")
plt.plot(x,y_pred3,label = "b = {}".format(b2),color="green")
plt.plot(x,y_pred,label = "b = 0 ",color = "blue")
plt.legend()
plt.show()
# green are overlab skyblue
```

```python
# a new experiment
b=-100
m=78.35
lr=0.1
```

```python
# plot GD using loop
epochs = 100
for i in range(epochs):

  loss_slop = -2*np.sum(m*x.ravel()-b)
  step_size = loss_slop*lr
  b=b-step_size

  y_pred = m*x+b

  plt.plot(x,y_pred)


plt.scatter(x,y)
plt.legend()
plt.show()
```
